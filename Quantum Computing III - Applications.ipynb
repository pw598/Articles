## The Four Postulates, Heavily Distilled

<h5>Postulate 1: State Space</h5>

<i>Associated to any physical system is a complex vector space known as the state space of the system. If the system is isolated, then the system is completely described by its state vector, which is a unit vector in the system's state space.</i>

For a qubit, the state space is a two-dimensional complex vector space. The condition that the state vector is a unit vector expresses the idea that the probabilities for the outcomes of a computational basis measurement add up to $1$.

<h5>Postulate 2: Unitary Dynamics</h5>

<i>The evolution of an isolated quantum system is described by a unitary matrix acting on the state space of the system.</i>

Our quantum gates demonstrate this. Recall that unitary matrices are the only ones which preserve length, and we want the quantum state to be normalized as such.

Most physical systems aren't actually isolated, but it turns out the dynamics of the atom are still closely described by a unitary transformation.

<h5>Postulate 3: Measurement</h5>

<i>Measurement in the computational basis governs how you get information.</i>

In the simplest version, measuring just a single qubit in the computational basis, a qubit in the state $\alpha \vert 0 \rangle + \beta \vert 1 \rangle$ will give outcome $0$ with probability $|\alpha|^2$ and outcome $1$ with probability $|\beta|^2$.

<h5>Postulate 4: Composite Systems</h5>

<i>The state space of a composite physical system is the tensor product of the state spaces of the component physical systems.</i>

If we have systems numbered $1$ through $n$, and system $j$ is in state $\vert \psi_j \rangle$, then the joint state of the total system is the tensor product of the individual states $\vert \psi_1 \rangle \otimes \vert \psi_2 \rangle \otimes \ldots \otimes \vert \psi_n \rangle$.

## Quantum Particles

Quantum particles follow the principles of quantum mechanics, a branch of physics concerned with phenomena at very small scales, such as atoms, molecules, and subatomic particles. A description of quantum particles involves:

<h5>Wave-Particle Density:</h5>

- Quantum particles have a dual nature in that they exhibit characteristics of both waves and particles. Depending on the experiment conducted, they can display wave-like behaviors such as interference patterns, or particle-like behaviors such as discrete energy levels.

<h5>Quantized Properties:</h5>

- Quantum particles often have quantized properties, meaning they can only possess certain discrete values. For instance, the energy levels of electrons in an atom are quantized, and they can only occupy specific allowed energy states.

<h5>Uncertainty Principle:</h5>

- Heisenberg's uncertainty principle states that there are inherent limitations to the precision with which we certain pairs of physical properties, such as position and momentum, can be simultaneously known.

<h5>Superposition:</h5>

- Quantum particles can exist in a state of superposition, where they simultaneously occupy multiple states until measured. This allows for complex behaviors and enables quantum systems for complex behaviors and enables quantum systems to perform multiple computations simultaneously.

<h5>Entanglement:</h5>

- Quantum particles can become entangled with each other, leading to correlations that persist even when the particles are separated by large distances.

<h5>Quantum Tunneling:</h5>

- Quantum particles can tunnel throudh energy barriers that would be impossible according to classical physics.

## Quantum Parallelism

Quantum parallelism is what all the hype is rightfully about. It is a key concept in quantum computing that allows quantum algorithms to perform computations on multiple inputs simultaneously. This phenomenon arises from the principles of superposition and entanglement in quantum mechanics. Some key applications include:

### Applications

<h5>Unsorted-Database Search:</h5>

- Grover's algorithm utilizes quantum parallelism to search an unsorted database of items in $O(\sqrt{N})$ time, compared to $O(N)$ time required by classical algorithms.

<h5>Factorization:</h5>

- Shor's algorithm exploits quantum parallelism to efficiently factorize large integers into their prime factors. This algorithm can factorize an $n$-bit integer in polynomial time, with a complexity estimated to be $O((log ~n)^3)$, compared to the sub-exponential complexity of the best-known classical algorithms.

<h5>Quantum Simulation:</h5>

- Quantum parallelism is essential for simulating quantum systems, which are inherently parallel in nature, such as molecules, materials, and condensed matter systems. This has implications on drug discovery, materials science, and chemistry.

<h5>Optimization and Machine Learning:</h5>

- Algorithms like Simon's algorithm and Shor's algorithm threaten to exploit the parallel evaluation of quantum states to reveal hidden structures in certain types of functions, which has implications toward breaking cryptographic schemes.

## Grover's Algorithm

<h5>1. Initialization:</h5>

- The quantum computer prepares a superposition of all possible states using Hadamard gates, which put all qubits into a superposition of both $0$ and $1$.

<h5>2. Oracle:</h5>

- An oracle gate, often represented $U_f$, marks the target state(s) in the search space, and flips the phase of the target state(s).

<h5>3. Amplitude Amplification:</h5>

- A sequence of operations that amplify the amplitude of the marked states and suppress the amplitude of the unmarked states. It consists of inverting amplitudes about the mean, by applying a series of gates including the oracle gate and Hadamard gate, and repeating this an optimal number of times.

<h5>4. Measurement:</h5>

- Finally, after a certain number of iterations, a measurement is performed on the qubits. This collapses the superposition into a single state, and with high probability it will be one of the marked states.

## Shor's Algorithm

Shor's algorithm efficiently factors large integers into their prime factors. It is groundbreaking because it demonstrates exponential speedup over the best-known classical algorithms for factoring, which become impractical for very large numbers. It therefore has potential to break encryption schemes based on such methods, in situations where a classical computer would fail to achieve this in a feasible timeframe.

<h5>1. Classical Preprocessing:</h5>

- Given the number $N$ that we want to factor, some classical computations are performed to ensure that $N$ is composite and not a power of a small prime.

<h5>2. Period Finding Using Quantum Fourier Transforms:</h5>

- Shor's algorithm fundamentally relies on finding a period of a function $f(x) = a^x ~mod ~N$, where $a$ is a randomly chosen integer <a href="https://en.wikipedia.org/wiki/Coprime_integers">coprime</a> to $N$.

- This involves a modular exponentiation operation controlled by a quantum register. This operation, represented as $U_f, maps \vert x \rangle \vert y \rangle$ to $\vert x \rangle \vert y \oplus f(x) \rangle$, where $\oplus$ represents addition modulo $N$.

- The QFT is then applied to find the period of $f(x)$. Quantum parallelism plays a significant role in exploring multiple periods simultaneously.

<h5>3. Classical Postprocessing:</h5>

- After obtaining a candidate period $r$, classical techniques such as <a href="https://en.wikipedia.org/wiki/Continued_fraction">continued fractions</a> are used to extract potential factors of $N$. If $r$ turns out to be even or $a^{r/2} ~mod ~N = -1$, then the algorithm restarts with a different $a$.

<h5>4. Factor Extraction:</h5>

- From the potential factors obtained in the previous step, greatest common divisor (GCD) calculations are performed to obtain the factors of $N$.

## Variational Quantum Eigensolver (VQE)

The VQE is a hybrid quantum-classical algorithm used to find approximate solutions to the eigenvalue problem, particularly in quantum chemistry and material science. Its goal is to find the lowest eigenvalue (ground state energy) and corresponding eigenvector (ground state wavefunction) of a given <a href="https://en.wikipedia.org/wiki/Hamiltonian_matrix">Hamiltonian</a>, which represents the energy operator of a quantum state.

<h5>Problem Setup:</h5>

- The VQE algorithm starts with a quantum system of interest, typically described by a <a href="https://en.wikipedia.org/wiki/Molecular_Hamiltonian">molecular Hamiltonian</a> in the context of quantum chemistry. This Hamiltonian represents the total energy of the system and is usually expressed in terms of Pauli matrices.

<h5>Parameterized Ansatz:</h5>

- The next step involves choosing a parameterized quantum circuit, known as ansatz, to prepare a trial state (an approximation of the ground truth state of the system). The parameters are adjusted during the optimization process to minimize the energy of the trial state.

<h5>Quantum Circuit Execution:</h5>

- The parameterized quantum circuit is implemented on a quantum computer to prepare the trial state. The quantum circuit typically includes operations such as single-qubit rotations and entangling gates to create entanglement and explore different configurations of the quantum state.

<h5>Expectation Value Measurement:</h5>

- After preparing the trial state, measurements are performed to estimate the expectation value of the Hamiltonian, which involves applying the Hamiltonian operator to the trial state and measuring the resulting energy. Since the Hamiltonian is usually expressed in terms of Pauli matrices, the expectation value can be estimated using Pauli measurements.

<h5>Classical Optimization:</h5>

- The estimated energy is then used as the objective function in a classical optimization algorithm, such as gradient descent or Nelder-Mead. The goal is to adjust the parameters of the ansatz circuit to minimize the energy, thereby finding an approximation to the ground state of the system.

<h5>Convergence:</h5>

- Steps 3 through 5 are repeated iteratively, until the optimization algorithm converges to a minimum within a specified tolerance.

<h5>Result Interpretation:</h5>

- The final set of parameters corresponds to an approximate ground state of the quantum system. The estimated energy provides an approximation to the true ground state energy, and the corresponding trial state can be analyzed to extract physical insights about the system.

### Applications of the VQE

The VQE has several applications across various fields:

<h5>Quantum Chemistry:</h5>

- VQE can be used to estimate the ground-state energy and properties of molecules, which is crucial for understanding chemical reactions, molecular structures, and electronic properties. VQE can aid in drug discovery, catalyst design, and materials development.

<h5>Quantum Optimization:</h5>

- VQE can be applied to solve optimization problems that arise in various domains such as machine learning. VQE can be integrated into quantum machine learning algorithms, providing speedups over classlical machine learning, particularly when involving quantum data or quantum systems.

<h5>Quantum Error Correction:</h5>

- By simulating noisy quantum systems and evaluating the performance of error correction codes, VQE can help in the development of fault tolerant quantum computers.
