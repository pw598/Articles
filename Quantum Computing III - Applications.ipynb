## The Four Postulates (Heavily) Distilled

### Postulate 1: State Space

<i>Associated to any physical system is a complex vector space known as the state space of the system. If the system is isolated, then the system is completely described by its state vector, which is a unit vector in the system's state space.</i>

For a qubit, the state space is a two-dimensional complex vector space. The condition that the state vector is a unit vector expresses the idea that the probabilities for the outcomes of a computational basis measurement add up to 1.

### Postulate 2: Unitary Dynamics

<i>The evolution of an isolated quantum system is described by a unitary matrix acting on the state space of the system.</i>

Our quantum gates demonstrate this. Recall that unitary matrices are the only ones which preserve length, and we want the quantum state to be normalized as such.

Most physical systems aren't actually isolated, but it turns out the dynamics of the atom are still closely described by a unitary transformation.

### Postulate 3: Measurement

<i>Measurement in the computational basis governs how you get information.</i>

In the simplest version measuring just a single qubit in the computational basis, a qubit in the state $\alpha \vert 0 \rangle + \beta \vert 1 \rangle$ will give outcome $0$ with probability $|\alpha|^2$ and outcome $1$ with probability $|\beta|^2$.

### Postulate 4: Composite Systems

The state space of a composite physical system is the tensor product of the state spaces of the component physical systems.

If we have systems numbered $1$ through $n$, and system $j$ is in state $\vert \psi_j \rangle$, then the joint state of the total system is the tensor product of the individual states $\vert \psi_1 \rangle \otimes \vert \psi_2 \rangle \otimes \ldots \otimes \vert \psi_n \rangle$.
