## History of Quantum Mechanics

Our understanding of quantum mechanics started developing in the early 1900s, to understand certain elements of behavior among atoms and subatomic particles which classical physics couldn't explain. To name some (though they weren't all known at the time):

- Wave-Particle duality
- Quantization of energy
- Atomic structure
- Quantum entanglement
- Quantum tunneling

### Early Contributors

Some of the major early contributions were:

- <u>Max Planck: The Planck Constant and Planck Relation (1900)</u>
    - Max Planck introduced the concept of quantization of energy, proposing that energy is not emitted or absorbed continuously, but rather, in discrete packets or 'quanta'. This led to development of the Planck constant, which quantifies the relationship between the energy of a photon and its frequency. The Planck relation, $E=hf$ (where $E$ is energy, $f$ is frequency, and $h$ is Planck's constant) became fundamental in understanding the behavior of particles at the quantum level.    

- <u>Louis de Broglie: The de Broglie Wavelength (1924)</u>
    - The de Broglie wavelength proposed that particles such as electrons can exhibit wave-like behavior. The concept of wave-particle duality revolutionized the understanding of the behavior of particles at the quantum level.

- <u>Erwin Schrodinger: The Schrodinger Equation (1925)</u>
    - Schrodinger's equation describes how the wave function of a physical system evolves over time. It has had a profound impact on quantum mechanics by allowing for the prediction of the probability distribution of a particle's position and momentum, as well as energy levels, within a quantum system.

- <u>Wolfgang Pauli: The Pauli Exclusion Principle (1925) and Pauli matrices (1927)</u>
    - The Pauli exclusion principle states that no two identical <a href="https://en.wikipedia.org/wiki/Fermion">fermion</a> particles can occupy the same quantum state simultaneously within a quantum system. The Pauli matrices are a set of three 2x2 complex (in one case) matrices used in quantum mechanics and quantum computing.

- <u>Werner Heisenberg: The Heisenberg Uncertainty Principle (1927)</u>
    - The Heisenberg uncertainty principle states that it is impossible to precisely measure both the position and momentum of a particle, challenging classical notions of determinism.

- <u>Paul Dirac: the Dirac Wave Equation (1928), Dirac Notation</u>
    - Paul Dirac's relativistic wave equation unified quantum mechanics and special relativity, providing a relativistic description of the behavior of electrons. It was used to predict the existence of antimatter. Dirac notation is a shorthand notation for basis vectors that is pervasive in quantum mechanics and quantum computing.

That's as far as I'll go beyond the scope of quantum computing, and into the more general aspects of quantum mechanics. There are many freely available, though math-heavy, books and papers on the subject.

## History of Quantum Computing

The theoretical foundations of quantum computing emerged in the 1980s, and proposed that by harnessing the principles of quantum mechanics, it would be possible to perform certain computations much more efficiently than classical computers.

In the 1990s, some groundbreaking algorithms were developed by researchers such as Peter Shor (Shor's algorithm, 1994) and Lov Grover (Grover's search algorithm, 1996), which demonstrated the theoretical potential of quantum computers to do tasks like search an unsorted database or factor large numbers with much greater efficiency than classical computers.

In the 2000s, various physical systems including trapped ions, superconducting qubits, and quantum dots were explored as candidates for building quantum processors, and the 2010s witnessed significant advancements in hardware and scalability. The field continues to grow rapidly, with ongoing efforts to overcome technical challenges like increasing qubit counts.

As the field continues to advance, quantum computing has the potential to bring about transformative changes in computing and scientific discovery. Quantum computers promise to solve problems that are intractable with current computing technologies, as their ability to compute multiple states simultaneously enables them to perform an indefinite number of tasks in parallel.

Unlike classical computers based on sequential information processing, quantum computing uses the properties of superposition, entanglement, and interference. But rather than increasing the available computing capacity, it reduces the capacity needed to solve a problem. Quantum computing uses three fundamental properties of quantum physics: 1) superposition, 2) interference, and 3) entanglement.

- <u>Superposition</u> is the quantum phenomenon by which a quantum system can exist in a complex linear combination of state 0 and state 1.
- <u>Interference</u> allows us to bias quantum systems toward the desired state
- <u>Entanglement</u> is the phenomenon by which entangled particles remain perfectly correlated even if separated by great distances

## Bits

Bits, zeros and ones, are the world's simplest alphabet, and can be strung together to encode any piece of information. In the early days of computing, computer scientists stored bits by making holes in paper cards. Other mediums developed over the years, an exhaustive list of which can be perused <a href="https://www.computerhistory.org/timeline/memory-storage/">here</a>.

Electrons orbit the nucleus of an atom, at specific distances from the nucleus called shells. If we have a way of detecting which shell an electron is in, we can pick two of these shells and use the location of the electron as a bit. If it is in a particularly chosen shell, it is $0$; otherwise, $1$.

## Classical States and Probability Vectors

Let's give the name $X$ to a system being considered, and use $\Sigma$ to refer to the set of classical states of $X$, let's say $X \in \{0,1\}$. Often, in information processing, our knowledge of $X$ is uncertain. We represent our knowledge of $X$ by assigning probabilities to each classical state in the set. Perhaps we believe there are equal probabilities of the two states in the example, $Pr(X=0)=0.5$ and $Pr(X=1)=0.5$. A more succint way to represent this is with a column vector,

$
\begin{bmatrix}
\frac{1}{2} \\
\frac{1}{2} \\
\end{bmatrix}
$

Note that as a matter of convention, we represent bit $0$ on top, bit $1$ on the bottom.

### Measuring Probablistic States

Consider what happens when we measure an item in a probablistic state; let's say a coin that is spinning too fast for us to tell how the confluence of deterministic classical physics will make it land. We would consider it to have $Pr(X=Heads)=0.5$ and $Pr(X=Tails)=0.5$. Once measured, a state is known, and the probability vector representing $X$ becomes a vector with $1$ on the entry corresponding to the bit being $0$, and $0$ for all other entries, indicating the classical state with certainty. The shorthand for these vectors is:

$
\vert 0 \rangle = 
\begin{bmatrix}
1 \\
0 \\
\end{bmatrix}
$
, 

pronounced "ket-zero", and 

$
\vert 1 \rangle = 
\begin{bmatrix}
0 \\
1 \\
\end{bmatrix}
$
, 

pronounced "ket-one". This is known as Dirac notation. Note that any two-dimensional vector can be expressed as a linear combination of these vectors. For example,

$
\frac{3}{4} \vert 0 \rangle + \frac{1}{4} \vert 1 \rangle = 
\begin{bmatrix}
\frac{3}{4} \\
\frac{1}{4} \\
\end{bmatrix}
$

We haven't yet said a word about qubits, nor left the realm of classical physics. But we have covered some fundamental background and notation used when discussing quantum computing, and hopefully these elements will now seem less mysterious when encountered in the quantum context, since we see that they generalize beyond that.

## Qubits and Quantum States

Just as classical computers rely on bits, quantum computers rely on quantum bits, called qubits. Like a bit, it has a state, which when measured yields a $0$ or $1$, however many more states (linear combinations of basis vectors) are achievable, and provide qubits with capabilities not available to classical bits.

A superposition is a linear combination of the basis vectors, e.g:


$0.6 \vert 0 \rangle + 0.8 \vert 1 \rangle = 0.6 \begin{bmatrix} 1 \\ 0 \\ \end{bmatrix} + 0.8 \begin{bmatrix} 0 \\ 1 \\ \end{bmatrix} = \begin{bmatrix} 0.6 \\ 0.8 \\ \end{bmatrix}$

There is a constraint that the squares of the amplitudes, the scalar coefficients of the vectors, must add up to $1$, as they do in this example.

$0.6^2 + 0.8^2 = 0.36 + 0.64 = 1$

This is the normalization constraint, and ensures the vector is a unit vector. We represent the amplitudes with $\alpha$ and $\beta$, and the equation generalizes to:

$
\vert \psi \rangle = \alpha \vert 0 \rangle + \beta \vert 1 \rangle = 
\begin{bmatrix}
\alpha \\
\beta \\
\end{bmatrix}
$

$s.t. |\alpha|^2 + |\beta|^2 = 1$

The quantum state is actually a two-dimensional vector in complex vector space; i.e., the elements can be complex numbers. But more on this later.

The qubit, when observed, has the probability of $\alpha^2$ of being in state $\vert 0 \rangle$ and probability $\beta^2$ of being in state $\vert 1 \rangle$.

In the two-dimensional qubit system, the top and bottom of the unit circle correspond to the basis vectors $\vert 0 \rangle$ and $\vert 1 \rangle$ respectively. The angle between the state vector $\vert 0 \rangle$ and $\vert \psi \rangle$, named $\theta$, controls the proximities of the vector head to the top and bottom of the system, thus containing the probabilities of state observed upon measurement. We can derive that:

$
\vert \psi \rangle = \alpha \vert 0 \rangle + \beta \vert 1 \rangle = 
\begin{bmatrix}
\alpha \\
\beta \\
\end{bmatrix}
$

is the same as:

$
\vert \psi \rangle = cos \frac{\theta}{2} \vert 0 \rangle + sin \frac{\theta}{2} \vert 1 \rangle = 
\begin{bmatrix}
cos \frac{\theta}{2} \\
sin \frac{\theta}{2} \\
\end{bmatrix}
$

<p>where $\theta$ is the angle between the standard basis vector, $\vert 0 \rangle = \begin{bmatrix} 1 \\ 0 \\ \end{bmatrix}$ and the qubit state vector, $\vert \psi \rangle$.</p>

## Logic Gates

Let's return to the classical realm for a moment. You are likely familiar with logic gates, mechanisms by which the state of an output bit is determined based on the state of one or more bits being input. The NOT gate, for example, outputs a $1$ if the input is $0$ and $0$ if the input is $1$.

Now, recall that we have the option to express even classical bits as probability vectors $\vert 0 \rangle$ and $\vert 1 \rangle$. To flip state from one basis vector to the other, we can pre-multiply by a matrix that is equivalent to the NOT gate, designated $X$ below.

$
X = 
\begin{bmatrix}
0 & 1 \\
1 & 0 \\
\end{bmatrix}
$

$X \vert 0 \rangle = \begin{bmatrix} 0 & 1 \\ 1 & 0 \\ \end{bmatrix} \begin{bmatrix} 1 \\ 0 \\ \end{bmatrix} = \begin{bmatrix} 0 \\ 1 \\ \end{bmatrix} = \vert 1 \rangle$

$X \vert 1 \rangle = \begin{bmatrix} 0 & 1 \\ 1 & 0 \\ \end{bmatrix} \begin{bmatrix} 1 \\ 0 \\ \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \\ \end{bmatrix} = \vert 0 \rangle$

Perhaps then it is not surprising that this is how a NOT (a.k.a. $X$) gate is implemented in the quantum realm, where the state of our qubit is always represented as a basis vector. 

This generalizes to a qubit in a state of superposition as follows:

$X(\alpha \vert 0 \rangle + \beta \vert 1 \rangle) = \begin{bmatrix} 0 & 1 \\ 1 & 0 \\ \end{bmatrix} \begin{bmatrix} \alpha \\ \beta \\ \end{bmatrix} = \begin{bmatrix} \beta \\ \alpha \\ \end{bmatrix} = \alpha \vert 1 \rangle + \beta \vert 0 \rangle$

Quantum logic gates can always be implemented as matrices, or combinations of matrices. Fittingly, the identity matrix from linear algebra leaves the input untransformed.

$
I = 
\begin{bmatrix}
1 & 0 \\
0 & 1 \\
\end{bmatrix}
$

$I \vert 0 \rangle = \begin{bmatrix} 1 & 0 \\ 0 & 1 \\ \end{bmatrix} \begin{bmatrix} 0 \\ 1 \\ \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \\ \end{bmatrix} = \vert 0 \rangle$

$I \vert 1 \rangle = \begin{bmatrix} 1 & 0 \\ 0 & 1 \\ \end{bmatrix} \begin{bmatrix} 0 \\ 1 \\ \end{bmatrix} = \begin{bmatrix} 0 \\ 1 \\ \end{bmatrix} = \vert 1 \rangle$

## Quantum Logic Gates

### The Pauli Matrices

The Pauli matrices are as follows:

$
    X = 
    \begin{bmatrix}
    0 & 1 \\
    1 & 0 \\
    \end{bmatrix}
    $

$
    Y = 
    \begin{bmatrix}
    0 & -i \\
    i & 0 \\
    \end{bmatrix}
    $

$
    Z = 
    \begin{bmatrix}
    1 & 0 \\
    0 & -1 \\
    \end{bmatrix}
    $

You'll notice $Y$ involves $-i$. These are indeed complex-values matrices, but more on that later. The $X$ (i.e., NOT) matrix behaves as discussed above.

- $X \vert 0 \rangle = \vert 1 \rangle$
- $X \vert 1 \rangle = \vert 0 \rangle$

The $Z$ matrix invokes a 'phase flip', having the action:

- $Z \vert 0 \rangle = \vert 0 \rangle$
- $Z \vert 1 \rangle = - \vert 1 \rangle$

### The Hadamard Gate

Denoted by $H$, the Hadamard gate does the following:

$H \vert 0 \rangle = \frac{\vert 0 \rangle + \vert 1 \rangle}{\sqrt{2}}$

$H \vert 1 \rangle = \frac{\vert 0 \rangle - \vert 1 \rangle}{\sqrt{2}}$

That is when $\vert 0 \rangle$ or $\vert 1 \rangle$ is the quantum state. This generalizes to:

$
H(\alpha \vert 0 \rangle + \beta \vert 1 \rangle) = 
\alpha \left( \frac{\vert 0 \rangle + \vert 1 \rangle}{\sqrt{2}} \right)
+
\beta \left( \frac{\vert 0 \rangle - \vert 1 \rangle}{\sqrt{2}} \right)
$

$
H(\alpha \vert 0 \rangle + \beta \vert 1 \rangle) = 
\frac{\alpha + \beta}{\sqrt{2}} \vert 0 \rangle + \frac{\alpha - \beta}{\sqrt{2}} \vert 1 \rangle
$

The matrix representation is:

$
H = \frac{1}{\sqrt{2}} 
\begin{bmatrix}
1 & 1 \\
1 & -1 \\
\end{bmatrix}
$

What is it's purpose? This operator allows us to move away from the basis state vectors $\vert 0 \rangle$ and $\vert 1 \rangle$, and puts the qubit into a balanced state of superposition. The Hadamard gate turns a qubit from state $\vert 0 \rangle$ into what is denoted $\vert + \rangle$ and from state $\vert 1 \rangle$ to what is denoted state $\vert - \rangle$, and it also reverses these transformations.

### The Bloch Sphere

The Bloch sphere represents a single qubit and its two basis states, which are located at opposite poles (north pole $= \vert 0 \rangle$, south pole $= \vert 1 \rangle$).

<img src="img/bloch.png" style="height: 300px; width:auto;">

he Hadamard gate places the qubit in a superposition state; specifically, a complex linear combination of the basis sates, which yields an equal probability of $0$ or $1$ when we measure the qubit.

The first superposition equation is as follows, and originates from the $\vert 0 \rangle$ state.

<h3>$\vert + \rangle = \frac{\vert 0 \rangle + \vert 1 \rangle}{\sqrt{2}}$</h3>

<h3>$\vert - \rangle = \frac{\vert 0 \rangle - \vert 1 \rangle}{\sqrt{2}}$</h3>

This is equal to a $\frac{\pi}{2}$ rotation about the X and Z axis of the Bloch sphere, counter-clockwise. When adding an $H$gate from the $\vert 0 \rangle$ basis state, the vector is placed on the positive x axis.

When initiating the qubit to the $\vert 1 \rangle$ state and then applying a Hadamard gate, the vector lies upon the negative x axis. The difference is referred to as a phase difference, which plays an important role in many quantum algorithms and in the topic of interference, though it's true that we use many shots (rounds of experimentation) in an experiment because quantum devices are not yet fault-tolerant.
